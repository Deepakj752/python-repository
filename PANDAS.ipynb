```python
df_stud['Percentage']=df_stud['marks']/200*100
df_stud
```

```python
# apply condition on percentage col
# we can create function for grade allocation
def grade(per):   # percentage col  assigned to per variable
    if (per>40):
        return 'A'
    elif (per<=40):
        return 'B'
    else:
        return 'C'

df_stud['Grade']=df_stud['Percentage'].apply(grade)    # dataframe[col].apply(functionname)       # functioname(arguments)
        
df_stud

```

```python
#value_counts--return total value avaiable in a col

df_stud['Grade'].value_counts()
```
